```python
notas = pd.read_csv('https://raw.githubusercontent.com/angelomaia/alura-cursos/master/Forma%C3%A7%C3%A3o%20Estat%C3%ADstica%20com%20Python/Data_Science-Testes_Estat%C3%ADsticos/ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby('movieId').mean()['rating']

nota_media_por_filme.head()
```

```python
ax = sns.histplot(nota_media_por_filme.values, kde=True)
ax.set(xlabel='Nota mÃ©dia', ylabel='FrequÃªncia')
ax.set_title('Notas mÃ©dias de filmes do Movielens 100k')
```

```python
quantidade_de_votas_por_filme = notas.groupby('movieId').count()
filmes_com_pelo_menos_10_votos = quantidade_de_votas_por_filme.query('rating >= 10').index
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
ax = sns.histplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values, kde=True)
ax.set(xlabel='Nota mÃ©dia', ylabel='FrequÃªncia')
ax.set_title('Notas mÃ©dias de filmes do Movielens 100k')
```

```python
ax = sns.boxplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel='Nota mÃ©dia', ylabel='FrequÃªncia')
ax.set_title('Notas mÃ©dias de filmes do Movielens 100k')
```

```python
ax = sns.histplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values,
                  kde=True,
                  cumulative=True)
ax.set(xlabel='Nota mÃ©dia', ylabel='FrequÃªncia')
ax.set_title('Notas mÃ©dias de filmes do Movielens 100k')
```
