## Utilizando Random Forest para todas as dimensões

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from numpy import random

SEED = 123321
random.seed(SEED)

valores_exames = data.drop(columns=['diagnostico','id'])
diagnostico = data.diagnostico
valores_exames_v1 = valores_exames.drop(columns='exame_33') # retirando coluna com valores NaN

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v1, diagnostico, 
                                                        train_size=0.7, random_state=SEED)

model = RandomForestClassifier(n_estimators=100)
model.fit(treino_x, treino_y)
print("A acurácia foi de %.2f%%" % (model.score(teste_x, teste_y) * 100))
```

### Testando com baseline (dummyclassifier)

```python
from sklearn.dummy import DummyClassifier

SEED = 123321
random.seed(SEED)

classificador_bobo = DummyClassifier(strategy='most_frequent')

classificador_bobo.fit(treino_x, treino_y)
print("A acurácia foi de %.2f%%" % (classificador_bobo.score(teste_x, teste_y) * 100))
```

```python
def grafico_violino(valores, inicio, fim):
  dados_plot = pd.concat([diagnostico, valores.iloc[:,inicio:fim]],axis=1)
  dados_plot = pd.melt(dados_plot, id_vars='diagnostico',var_name='exames',value_name='valores')

  plt.figure(figsize=(15,10))
  sns.violinplot(x='exames',y='valores',
               hue='diagnostico',
               data=dados_plot,
               split = 2)

  plt.xticks(rotation=90)

grafico_violino(valores_exames_v2,10,21)
```

```python
valores_exames_v3 = valores_exames_v2.drop(columns=['exame_29','exame_4'])

def classificar(valores):
  SEED = 123321
  random.seed(SEED)
  treino_x, teste_x, treino_y, teste_y = train_test_split(valores, diagnostico, 
                                                          train_size=0.7, random_state=SEED)

  model = RandomForestClassifier(n_estimators=100)
  model.fit(treino_x, treino_y)
  print("A acurácia foi de %.2f%%" % (model.score(teste_x, teste_y) * 100))
```

```python
# Selecionar as K melhores features

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

selectkbest = SelectKBest(chi2, k = 10)

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v1.drop(columns=['exame_3',
                                                                                        'exame_24',
                                                                                        'exame_29',
                                                                                        'exame_4',
                                                                                        'exame_11',
                                                                                        'exame_16',
                                                                                        'exame_20',
                                                                                        'exame_10']), 
                                                        diagnostico,
                                                        train_size=0.7, 
                                                        random_state=SEED)

selectkbest.fit(treino_x, treino_y)

treino_kbest = selectkbest.transform(treino_x)
teste_kbest = selectkbest.transform(teste_x)

classificador = RandomForestClassifier(n_estimators = 200, random_state=SEED)
classificador.fit(treino_kbest, treino_y)
print("A acurácia foi de %.2f%%" % (classificador.score(teste_kbest, teste_y) * 100))
```

```python
from sklearn.metrics import confusion_matrix

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_kbest))

matriz_confusao
```

```python
plt.figure(figsize = (10,8))
sns.set(font_scale=2)
sns.heatmap(matriz_confusao, 
            annot = True, 
            fmt = "d", 
            cmap='viridis').set(xlabel = "Predicted", 
                                ylabel = "Real")
```

## Feature Selection por Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v1.drop(columns=['exame_3',
                                                                                        'exame_24',
                                                                                        'exame_29',
                                                                                        'exame_4',
                                                                                        'exame_11',
                                                                                        'exame_16',
                                                                                        'exame_20',
                                                                                        'exame_10']), 
                                                        diagnostico,
                                                        train_size=0.7, 
                                                        random_state=SEED)

classificador = RandomForestClassifier(n_estimators = 200, random_state=SEED)
classificador.fit(treino_x, treino_y)

rfe_select = RFE(estimator = classificador, n_features_to_select = 10, step = 1)
rfe_select.fit(treino_x, treino_y)
treino_rfe = rfe_select.transform(treino_x)
teste_rfe = rfe_select.transform(teste_x)

classificador.fit(treino_rfe, treino_y)

print("A acurácia foi de %.2f%%" % (classificador.score(teste_rfe, teste_y) * 100))
```

```python
from sklearn.metrics import confusion_matrix

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_rfe))

matriz_confusao

plt.figure(figsize = (10,8))
sns.set(font_scale=2)
sns.heatmap(matriz_confusao, 
            annot = True, 
            fmt = "d", 
            cmap='viridis').set(xlabel = "Predicted", 
                                ylabel = "Real")
```

## Feature Selection com RFE Cross-validation (RFECV)

```python
from sklearn.feature_selection import RFECV

treino_x, teste_x, treino_y, teste_y = train_test_split(valores_exames_v1.drop(columns=['exame_3',
                                                                                        'exame_24',
                                                                                        'exame_29',
                                                                                        'exame_4',
                                                                                        'exame_11',
                                                                                        'exame_16',
                                                                                        'exame_20',
                                                                                        'exame_10']), 
                                                        diagnostico,
                                                        train_size=0.7, 
                                                        random_state=SEED)

classificador = RandomForestClassifier(n_estimators = 200, random_state=SEED)
classificador.fit(treino_x, treino_y)

rfecv_select = RFECV(estimator = classificador, cv = 5, step = 1, scoring="accuracy") # Setando o algoritmo de RFECV
rfecv_select.fit(treino_x, treino_y)
treino_rfecv = rfecv_select.transform(treino_x)
teste_rfecv = rfecv_select.transform(teste_x)

classificador.fit(treino_rfecv, treino_y)

print("A acurácia foi de %.2f%%" % (classificador.score(teste_rfecv, teste_y) * 100))
```

```python
from sklearn.metrics import confusion_matrix

matriz_confusao = confusion_matrix(teste_y, classificador.predict(teste_rfecv))

matriz_confusao

plt.figure(figsize = (10,8))
sns.set(font_scale=2)
sns.heatmap(matriz_confusao, 
            annot = True, 
            fmt = "d", 
            cmap='viridis').set(xlabel = "Predicted", 
                                ylabel = "Real")
```
